```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as colors
import seaborn as sns
import itertools
from scipy.stats import norm
import scipy.stats
from sklearn.naive_bayes import GaussianNB
```

```python
# Open the file
file_path = "datacum.txt"
with open(file_path, "r") as file:
    lines = file.readlines()

# Initialize variables to store the data
data_groups = []
status_groups = []
# Loop through the lines
for line in lines:
    # Check if the line starts with "#####"
    if not(line.startswith("#####")):
        # Otherwise, parse the line and append it to the current group
        attributes = line.strip().split(",")
        #Get Status
        status = [int(attr) for attr in attributes[1:2]]
        status_groups.append(status)
        # only interested in the attributes 3 to 11
        data_point = [int(attr) for attr in attributes[2:11]]
        data_groups.append(data_point)

# remove umpty array
data_groups = [group for group in data_groups if group]

# Print the number of groups and the first few data points of each group
print(data_groups)
```

```python
iris = pd.DataFrame(data_groups, columns=['Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli', 'Mitoses'])
```

```python
def predict_cancer(data_groups):
    # Define weights
    w2 = (3, 0, 2, 2, 0, 2, 2, 1, 3)

    # Calculate scores
    scores_list = []
    for data in data_groups:
        # Ensure that the length of data and w2 is the same
        if len(data) != len(w2):
            print("Error: Length mismatch between data and w2")
            break
        # Calculate score for each data point
        score = sum(w * x for w, x in zip(w2, data))

        scores_list.append(score)

    return scores_list
```

```python
# Call the function with the entire data_groups list
predictions = predict_cancer(data_groups)
print("Predictions:", predictions)
```

```python
# remove umpty array
status_groups = [group for group in status_groups if group]
print(status_groups)

predictions_result = []
for i in  predictions:
    if i >= 50:
        predictions_result.append([4])
    else:
        predictions_result.append([2])
predictions = predict_cancer(data_groups)
print( predictions_result)
```

```python
#Function equal result with class
def similarity_percentage(array1, array2):
    if len(array1) != len(array2):
        raise ValueError("%:")
    
    total_elements = len(array1)
    same_elements = sum(1 for a, b in zip(array1, array2) if a == b)
    
    return (same_elements / total_elements) * 100

# Use function similarity_percentage
percent_similarity = similarity_percentage(status_groups, predictions_result)
print("%:", percent_similarity)
```
