```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
# Change to data path on your computer
data =pd.read_csv("E:\\machine\\machinebegin\\beginmachine\\week2\\SAT_GPA.csv")
# Show the description of data
data.describe()
# Set to training data (x, y)
y = data['GPA'].values.reshape(-1,1)
x = data['SAT'].values.reshape(-1,1)
# Split the data set

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 42)

# Remind that we need to put component x_0 = 1 to x
plt.scatter(x,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Get the coefficients of the regression line
intercept = model.intercept_[0]
slope =  model.coef_[0][0]

plt.scatter(x_train, y_train)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.plot(x_train, slope * x_train + intercept, color = 'red')
plt.show
```

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on Validation Data:", mse)
```
